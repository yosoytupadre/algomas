# INSERTION SORT

```python
import numpy as np
import random
from itertools import permutations
import matplotlib.pyplot as plt
```

## PASOS, COMPARACIONES E INTERCAMBIOS

```python
def insertionSortCount(aList):
    steps = 0
    comps = 0
    swaps = 0
    i = 1
    
    while(i<len(aList)):
        temp = aList[i]
        j=i-1
        while(True):
            comps+=1
            if(j<0):break
            if(temp>=aList[j]):break
            swaps+=1
            aList[j+1]=aList[j]
            j-=1;steps+=3
        aList[j+1]=temp;steps+=1;
        i+=1;steps+=4
    steps+=1
    
    return steps,comps,swaps;
```

## PERMUTACIONES ITERTOOLS 

### TAMAÃ‘O PERMUTACION

```python
n = 3
r = n
s = random.sample(range(1,r+1), n)
print s
```

```python
x = list(permutations(s, n))

totalSteps = np.empty(len(x))
totalComps = np.empty(len(x))
totalSwaps = np.empty(len(x))
for i in range(len(x)):
    steps,comps,swaps = insertionSortCount(list(x[i]))
    totalSteps[i]=steps
    totalComps[i]=comps
    totalSwaps[i]=swaps
```

### Pasos

```python
minSteps = min(totalSteps)
maxSteps = max(totalSteps)
plt.hist(totalSteps, bins=np.arange(minSteps, maxSteps+2, 1), normed=1);
print("\n Numero promedio de instrucciones: "+str(np.average(totalSteps)))
```

### Comparaciones

```python
minComps = min(totalComps)
maxComps = max(totalComps)
plt.hist(totalComps, bins=np.arange(minComps, maxComps+2, 1), normed=1);
print("\n Numero promedio de comparaciones: "+str(np.average(totalComps)))
```

### Swaps

```python
minSwaps = min(totalSwaps)
maxSwaps = max(totalSwaps)
plt.hist(totalSwaps, bins=np.arange(minSwaps, maxSwaps+2, 1), normed=1);
print("\n Numero promedio de swaps: "+str(np.average(totalSwaps)))
```
