# QUANTOPIAN

## PYTHON

```python
from time import time
import matplotlib.pyplot as plt

def multMatrix(min,max,retry):
    tMeans=[0.0]*(max-min)
    xAxis = range (min,max)
    
    for T in range (min,max):
        # Creacion de la matriz de 1s y 2s
        matriz1 = []
        matriz2 = []
        matriz3 = []

        for i in range(T):
            matriz1.append([])
            matriz2.append([])
            matriz3.append([])
            for j in range(T):
                matriz1[i].append(1)
                matriz2[i].append(2)
                matriz3[i].append(0)

        # Multiplicacion de matrices
        tMean = 0.0
        oper = ((T*T*T) + (T*T*(T-1)))
        for _ in range(retry):
            t0 = time()
            for i in range(T):
                    for j in range(T):
                        for k in range(T):
                            matriz3[i][j] += matriz1[i][k] * matriz2[k][j]
            tf = time()
            tt = (tf-t0)
            tMean += (tt/oper)
        
        # Tiempo promedio i
        tMeans[T-min] = (tMean/retry)
    
    # Tiempo promedio total
    tMean=((sum(tMeans))/(len(tMeans)))
    printer(min,max,tMean,tMeans)
    return tMeans
```

```python
QTp1 = multMatrix(2,21,20)
```

```python
QTp2 = multMatrix(2,101,10)
```

```python
QTp3 = multMatrix(20,101,20)
```

```python
QTp4 = multMatrix(20,51,10)
```

## PYTHON NUMPY

```python
import numpy as np
from time import time
import matplotlib.pyplot as plt

def multMatrixNumpy(min,max,retry):
    tMeans=[0.0]*(max-min)
    
    for T in range (min,max):
        # Creacion de la matriz de 1s y 2s
        matriz1 = np.ones((T, T))
        matriz2 = np.ones((T, T))*2
        matriz3 = np.zeros((T, T))

        # Multiplicacion de matrices
        tMean = 0.0
        oper = ((T*T*T) + (T*T*(T-1)))
        for _ in range(retry):
            t0 = time()
            matriz3 = matriz1 * matriz2
            tf = time()
            tt = (tf-t0)
            tMean += (tt/oper)

        # Tiempo promedio i
        tMeans[T-min] = (tMean/retry)
    
    # Tiempo promedio total
    tMean=((sum(tMeans))/(len(tMeans)))
    
    printer(min,max,tMean,tMeans)
    return tMeans
```

```python
QTnp1 = multMatrixNumpy(2,21,20)
```

```python
QTnp2 = multMatrixNumpy(2,101,10)
```

```python
QTnp3 = multMatrixNumpy(20,101,20)
```

```python
QTnp4 = multMatrixNumpy(20,51,10)
```
